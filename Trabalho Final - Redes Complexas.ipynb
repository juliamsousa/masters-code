```python
# Importando pacote timer para medir tempo de execuÃ§Ã£o
from timeit import default_timer as timer
# Importando o pacote da ctgan
from ctgan import CTGAN
# Importando o dataset de demonstraÃ§Ã£o para primeiros testes
from ctgan import load_demo

# Carregando os dados de demonstraÃ§Ã£o em uma variÃ¡vel
real_data = load_demo()

# Descrevendo os dados discretos do dataset
discrete_columns = [
    'workclass',
    'education',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'native-country',
    'income'
]

# Instanciando um modelo da CTGAN com 10 Ã©pocas para aprendizado
ctgan = CTGAN(epochs=10)
# Treinando o modelo com os "dados reais" informando as colunas de dados discretos
ctgan.fit(real_data, discrete_columns)

#### Registra o tempo de inÃ­cio da execuÃ§Ã£o do treinamento
start_time = timer()
# CriaÃ§Ã£o dos dados sintÃ©ticos
synthetic_data = ctgan.sample(1000)
#### Registra o tempo de fim de execuÃ§Ã£o do treinamento
end_time = timer()
#### Calcula o tempo decorrido
elapsed_time = end_time - start_time

print(elapsed_time)
```
